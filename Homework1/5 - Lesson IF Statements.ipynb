#### Demo 2

```python
price = int(input("What is your cellular price ? : "))

if price >= 0 and price <=2000:
    print('Very cheap!')
elif price >= 2001 and price <=3000:
    print('cheap')
elif price >= 3001 and price <=4000:
    print('medium')
elif price >= 4001 and price <=5000:
    print('Expensive')
else:
    print('Apple')
```

This could also be written as : 

```python
price = int(input("What is your cellular price ? : "))

if 0 <= price <= 2000:
    print('Very cheap!')
elif 2001 <= price <= 3000:
    print('cheap')
elif 3001 <= price <= 4000:
    print('medium')
elif 4001 <= price <= 5000:
    print('Expensive')
else:
    print('Apple')
```

Nested IF

```python
price = int(input("What is your cellular price ? Please type a number between 0-7000 : "))

if 0 <= price <= 7000:
    if 0 <= price <= 2000:
        print('Very cheap!')
    elif 2001 <= price <= 3000:
        print('cheap')
    elif 3001 <= price <= 4000:
        print('medium')
    elif 4001 <= price <= 5000:
        print('Expensive')
    else:
        print('Apple')
else:
    print('You\'ve typed a number out of the range of 0-7000')
```

### Using `and` / `or`

#### Demo 1

Create a program that takes a 2 digit number and decides whether it can be divided by it`s digits

```python
v_number = 82 
print(v_number%10)
print(v_number//10)
```

```python
v_num = int(input('Please type a number between 10-99 : ')) # 17
digit_1 = v_num//10 #=> 1
digit_2 = v_num%10  #=> 7

if v_num % digit_1 == 0 and v_num % digit_2 == 0: #36 for example 
    print('The number {} can be divided by both {} and {}'.format(v_num,digit_1,digit_2))
elif v_num % digit_1 == 0: # 13 for example
    print('The number {} can only be divided by {} '.format(v_num,digit_1))
elif v_num % digit_2 == 0: # 35 for example
    print('The number {} can only be divided by {} '.format(v_num,digit_2))
else: # 27 for example
    print('The number {} cannot be divided by either {} or {}'.format(v_num,digit_1,digit_2))
```

Nested IF

```python
v_num = int(input('Please type a number between 10-99 : '))
digit_1 = v_num//10 #=> 1
digit_2 = v_num%10  #=> 7

if 10 <= v_num <= 99:
    if v_num % digit_1 == 0 and v_num % digit_2 == 0: #36 for example 
        print('The number {} can be divided by both {} and {}'.format(v_num,digit_1,digit_2))
    elif v_num % digit_1 == 0: # 13 for example
        print('The number {} can only be divided by {} '.format(v_num,digit_1))
    elif v_num % digit_2 == 0: # 35 for example
        print('The number {} can only be divided by {} '.format(v_num,digit_2))
    else: # 27 for example
        print('The number {} cannot be divided by either {} or {}'.format(v_num,digit_1,digit_2))
else:
    print('The number is not in the range of 10-99')
```
